```python
from bs4 import BeautifulSoup
import pandas as pd
import requests as req
import numpy as np
```

```python
URL="https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1675240745&sprefix=ba%2Caps%2C283&ref=sr_pg_1"
```

```python
My_user_agent=""
```

```python
HEADERS={'User-Agent':My_user_agent,'Accept-Language':'en-US,en;q=0.5'}
```

```python
def getnextpage(soup,url):
    # this will return the next page URL
	webpage = req.get(url, headers=HEADERS)
	soup = BeautifulSoup(webpage.content, "html.parser")

	link=soup.find('a', {'class': 's-pagination-item s-pagination-next s-pagination-button s-pagination-separator'})
	if link:
		link=link.get("href")
		url = 'https://www.amazon.in' + str(link)
		return url
	else:
		return

def get_details(url):
    webpage = req.get(URL, headers=HEADERS)
    soup = BeautifulSoup(webpage.content, "html.parser")
    divs = soup.find_all("div", attrs={'class':'sg-col-20-of-24 s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 sg-col s-widget-spacing-small sg-col-12-of-16'})
    for div in divs:
        l=div.find('a',attrs={'class':'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})
        li=l.get("href")
        Links_list.append("https://www.amazon.in"+str(li))
        name=l.find("span")
        Names_list.append(name.text)
        ratings=div.find("span",attrs={"class":"a-icon-alt"})
        if ratings!=None:
            Rating_list.append(ratings.text)
        else:
            Rating_list.append(np.nan)
        price=div.find("span",attrs={"class":"a-offscreen"})
        Price_list.append(price.text)
        reviews=div.find("span",attrs={"class":"a-size-base s-underline-text"})
        if reviews!=None:
            Review_list.append(reviews.text)
        else:
            Review_list.append(np.nan)
        

    
Links_list = []
Names_list = []
Price_list = []
Rating_list = []
Review_list = []

webpage = req.get(URL, headers=HEADERS)
soup = BeautifulSoup(webpage.content, "html.parser")
while True:
	get_details(URL)
	print(URL)
	url = getnextpage(soup,URL)
	if not url:
		break
	URL=url
```

```python
df=pd.DataFrame(list(zip(Links_list,Names_list,Price_list,Rating_list,Review_list)),columns=['PRODUCT_URL',
'PRODUCT_NAME','PRICE','RATINGS','REVIEWS'])
df
```

```python
df.to_csv("amazon_data.csv", header=True, index=False)
```

## Generating Info for Every Product

```python
description=[]
Asin_list=[]
Manufacturer_list=[]
products_links=[]
count=1
for product_url in Links_list[:201]:    
    webpage = req.get(product_url, headers=HEADERS)
    soup = BeautifulSoup(webpage.content, "html.parser")
    div = soup.find("div", attrs={'id':'detailBullets_feature_div'})
    ul=None
    if div!=None:
        ul=div.find("ul", attrs={'class':'a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list'})
        details=ul.find_all("span",attrs={"class":"a-list-item"})
        found=False
        s=""
        for l in details:
            result=l.find_all("span")
            title=str(result[0].text)
            index=title.find("\n")
            if title[:index]=="ASIN":
                Asin_list.append(result[1].text)
            elif (title[:index]=="Manufacturer") and found==False:
                Manufacturer_list.append(result[1].text)
                found=True
            s+=title[:index]+" : "+result[1].text+" , "
        description.append(s)
    else:
        Asin_list.append(np.nan)
        Manufacturer_list.append(np.nan)
        description.append(np.nan)
    print(str(count)+" "+product_url)
    products_links.append(product_url)
    count+=1
```

```python
df=pd.DataFrame(list(zip(products_links,Asin_list,Manufacturer_list,description)),columns=['URL','ASIN','Manufacturer','Description'])
df
```

```python
df.to_csv("Product_details.csv",header=True,index=False)
```
